```python
import os
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from string import punctuation

index = {}
nltk.download("stopwords")
nltk.download("punkt")


def lower(text):
    # return "".join(char.lower() for char in text if char.isalpha())
    return text.lower()
def token(text):
    return word_tokenize(text)

def stopword(tokens):
    stopwords_set = set(stopwords.words("english"))
    filtered_sentence = [w for w in tokens if w not in stopwords_set]
    return filtered_sentence

def remove_punctuation(input_string):
    cleaned_string = ""
    for char in input_string:
        # Check if the character is not a punctuation character
        if char not in string.punctuation:
            cleaned_string += char
    return cleaned_string

def remove_spaces(tokens):
    return [token for token in tokens if token.strip()]

def save_preprocessed_text(texts, output_folder, filename):
    output_path = os.path.join(output_folder, filename)
    with open(output_path, 'w') as output_file:
        output_file.write(" ".join(texts))

def convert(text):
            original_text = text  # Store the original text for later use if needed
            text = lower(text)
            text = remove_punctuation(text)
            tokens = token(text)
            tokens = remove_spaces(tokens)
            tokens = stopword(tokens)
            return tokens

def add_document(doc_id, tokens):
        for position, term in enumerate(tokens):
            if term not in index:
                index[term] = {}
            if doc_id not in index[term]:
                index[term][doc_id] = []
            index[term][doc_id].append(position)

def build_index_from_files(folder_path):
   for file_name in (os.listdir(folder_path)):
        file =file_name
        number_str = ''.join(char for char in file_name if char.isdigit())
        doc_id = int(number_str)
        # print (file_name)
        # print(doc_id)
        file_path = os.path.join(folder_path, file_name)
        if os.path.isdir(file_path):
            continue
        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                # document = file.read()
                # terms = document.split()
                # add_document(doc_id, tokens)
                content = file.read()
                tokens = tokenize(content)
                add_document(doc_id, tokens)

        except FileNotFoundError:
            print(f"File not found: {file_path}")

def search(query_terms):
        # Initialize the result with all documents
        result = set(range(len(index)))

        for term in query_terms:
            if term in index:
                # Intersect the result with the documents containing the current term
                result &= set(index[term].keys())

        return result
def tokenize(text):
        # A simple tokenizer that splits text into words
        words = []
        current_word = ""
        for char in text.lower():
            if char.isalnum():
                current_word += char
            else:
                if current_word:
                    words.append(current_word)
                    current_word = ""
        if current_word:
            words.append(current_word)
        return words

# Example usage:
folder_path = r'/content/drive/MyDrive/Preprocessed_Files'
# positional_index = UserDefinedPositionalIndex()

# Building the positional index from files
build_index_from_files(folder_path)
sorted_terms = sorted(index.keys())
for term in sorted_terms:
   print(f"Term: {term}")
   doc_positions = index[term]
   for doc_id, positions in doc_positions.items():
       print(f"  Document ID: {doc_id}, Positions: {positions}")
```

```python
def search_phrase(index, terms):
    # terms = phrase_query.split()
    if len(terms) < 2:
        raise ValueError("Phrase queries must contain at least two terms.")

    # Initialize result set with positions of the first term
    result_set = set(index[terms[0]])
    for term in terms:
        term_positions = index[term]
        result_set.intersection_update(set(term_positions.keys()))
    result_copy = set(result_set)  # Create a copy of the set
    lst=[]
    for doc_id in result_copy:

      positions = [set(index[term][doc_id]) for term in terms]
      positions_1 = positions[0]
      positions_2 = positions[1]
      if any(position2 - 1 == position1 for position1 in positions_1 for position2 in positions_2):
          lst.append(doc_id)
    return lst

# Example Usage:
phrase_query = "gig is a bag"
con = convert(phrase_query)
print(f"The result for the query:- {phrase_query}")
result_documents = search_phrase(positional, con)

print(f"Documents containing the phrase '{phrase_query}': {result_documents}")
```
